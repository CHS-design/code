# 深度学习


## 0. 深度学习杂项

### 0.1 深度学习 Deep Learning

深度学习是机器学习的一个重要方向，核心是用多层神经网络从数据中自动学习特征表示。

传统机器学习通常需要人工设计特征，例如房屋面积、房间数量、纹理统计等；深度学习更强调让模型通过多层计算自己学习特征。层数越深，模型越容易从简单特征组合出更抽象的特征。

### 0.2 直观理解

- 浅层网络：更容易学习边缘、颜色、简单数值关系等低层特征。
- 深层网络：可以组合出物体局部、语义关系、复杂模式等高层特征。
- 训练过程：用数据计算预测误差，再不断调整参数，让预测结果接近真实标签。

### 0.3 常见应用

| 方向 | 示例 |
| --- | --- |
| 图像识别 | 分类、目标检测、医学影像分析 |
| 语音处理 | 语音识别、声纹识别 |
| 自然语言处理 | 文本分类、机器翻译、问答系统 |
| 推荐系统 | 用户兴趣建模、点击率预测 |

深度学习的优势是表达能力强，缺点是通常需要更多数据、算力和调参经验，也更容易出现过拟合、梯度不稳定等问题。

## 1. 神经元和层 Neuron and Layer

### 1.1 神经元 Neuron

神经元是神经网络中的基本计算单元。它接收输入，对输入做加权求和，再加上偏置，最后通过激活函数得到输出。

单个神经元的计算为：

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

$$
a = g(z)
$$

其中：

- $x_1, x_2, \ldots, x_n$：输入特征
- $w_1, w_2, \ldots, w_n$：权重参数
- $b$：偏置参数
- $z$：线性加权结果
- $g(z)$：激活函数
- $a$：神经元输出，也叫激活值

### 1.2 层 Layer

层是由多个神经元组成的计算结构。同一层中的神经元通常接收上一层的输出，并分别完成线性计算和非线性变换。

第 $l$ 层的矩阵形式为：

$$
\mathbf{z}^{[l]} = \mathbf{W}^{[l]}\mathbf{a}^{[l-1]} + \mathbf{b}^{[l]}
$$

$$
\mathbf{a}^{[l]} = g(\mathbf{z}^{[l]})
$$

其中：

- $\mathbf{a}^{[l-1]}$：上一层输出
- $\mathbf{W}^{[l]}$：第 $l$ 层权重矩阵
- $\mathbf{b}^{[l]}$：第 $l$ 层偏置向量
- $\mathbf{z}^{[l]}$：第 $l$ 层线性输出
- $\mathbf{a}^{[l]}$：第 $l$ 层激活输出

### 1.3 常见层类型

| 层类型 | 作用 | 常见场景 |
| --- | --- | --- |
| 全连接层 Fully Connected | 对输入向量做整体特征变换 | 表格数据、分类器尾部 |
| 卷积层 Convolutional | 提取局部空间特征 | 图像、视频 |
| 池化层 Pooling | 降低特征图尺寸，保留主要信息 | CNN 中间层 |
| 输出层 Output | 生成最终预测结果 | 回归、分类 |

### 1.4 注意点

- $z$ 是线性部分的结果，$a$ 是经过激活函数后的结果。
- 如果没有激活函数，多层线性变换叠加后仍接近一个线性变换，模型难以表示复杂关系。
- 层数和神经元数量越多，表达能力越强，但也更容易过拟合，训练成本更高。

## 2. 神经网络 Neural Network

### 2.1 定义

神经网络是由输入层、隐藏层和输出层组成的模型。它通过多层参数变换，把输入数据映射为预测结果。

| 结构 | 作用 |
| --- | --- |
| 输入层 Input Layer | 接收原始特征或数据表示 |
| 隐藏层 Hidden Layer | 提取、组合和变换特征 |
| 输出层 Output Layer | 给出最终预测结果 |

多层感知机（Multilayer Perceptron, MLP）是常见的前馈神经网络。实际应用中还会用到 CNN、RNN、Transformer 等结构。

### 2.2 前向传播 Forward Propagation

前向传播是用当前参数完成一次预测：输入从输入层开始，依次经过隐藏层和输出层，最终得到预测值。

以第 $l$ 层为例：

$$
\mathbf{z}^{[l]} = \mathbf{W}^{[l]}\mathbf{a}^{[l-1]} + \mathbf{b}^{[l]}
$$

$$
\mathbf{a}^{[l]} = g(\mathbf{z}^{[l]})
$$

### 2.3 损失计算 Loss

模型预测值 $\hat{y}$ 与真实值 $y$ 之间的差距由损失函数表示。训练时希望损失越小越好。

$$
L = \text{loss}(y, \hat{y})
$$

常见搭配：

| 任务 | 输出形式 | 常用损失 |
| --- | --- | --- |
| 回归 | 连续值 | MSE、MAE |
| 二分类 | 一个概率 | Binary Cross-Entropy |
| 多分类 | 多类别概率分布 | Cross-Entropy |

### 2.4 反向传播 Backpropagation

反向传播用于计算损失函数对各层参数的梯度。它从输出层开始，把误差信息逐层传回前面的层，并利用链式法则计算每个参数该如何调整。

训练流程可以概括为：

1. 前向传播，得到预测值。
2. 计算损失，衡量预测值和真实值的差距。
3. 反向传播，计算各参数梯度。
4. 使用优化器更新权重和偏置。
5. 重复训练，直到损失下降到较合理的水平。

### 2.5 训练目标

神经网络训练的目标是找到一组参数，使模型在训练集和未见过的数据上都有较好的预测效果。

需要同时关注：

- 训练损失是否下降。
- 验证集效果是否提升。
- 是否出现过拟合或欠拟合。
- 学习率、批大小、网络深度是否合适。

## 3. 全连接神经网络 Fully Connected Neural Network

### 3.1 定义

全连接神经网络是最基础的一类神经网络结构。所谓全连接，是指相邻两层之间的神经元两两相连。

也就是说，上一层的每一个输出都会参与下一层每一个神经元的计算。

### 3.2 结构

一个简单的全连接网络可以表示为：

$$
\mathbf{x} \rightarrow \text{Hidden Layer} \rightarrow \text{Output Layer} \rightarrow \hat{y}
$$

其中：

- $\mathbf{x}$：输入特征向量
- Hidden Layer：隐藏层，用于特征变换
- Output Layer：输出层，用于得到预测结果
- $\hat{y}$：模型预测值

### 3.3 计算过程

全连接层的核心计算形式是：

$$
\mathbf{z} = \mathbf{W}\mathbf{x} + \mathbf{b}
$$

$$
\mathbf{a} = g(\mathbf{z})
$$

其中：

- $\mathbf{W}$：权重矩阵
- $\mathbf{x}$：输入向量
- $\mathbf{b}$：偏置向量
- $g$：激活函数
- $\mathbf{a}$：该层输出

### 3.4 参数量

如果输入维度为 $d$，输出神经元数量为 $h$，则这一层参数量为：

$$
d \times h + h
$$

其中 $d \times h$ 是权重数量，$h$ 是偏置数量。输入维度很高时，全连接层参数量会迅速增加。

### 3.5 特点

优点：

- 结构简单，容易理解。
- 适合处理已经整理成向量形式的数据。
- 常作为分类或回归模型的最后几层。

缺点：

- 参数量可能很大，容易过拟合。
- 对图像、序列等具有空间或时间结构的数据利用不充分。
- 输入维度很高时，计算和存储成本都会增加。

## 4. 激活函数 Activation Function

### 4.1 定义

激活函数用于给神经网络引入非线性。没有激活函数时，多层线性变换叠加后仍相当于一个线性变换，模型表达能力会很有限。

### 4.2 Sigmoid 函数

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

特点：

- 输出范围是 $(0,1)$。
- 常用于二分类任务的输出层。
- 输出可以理解为属于正类的概率。

缺点：

- 当输入很大或很小时，梯度会接近 0，容易出现梯度消失。
- 输出不是以 0 为中心，隐藏层中一般不优先使用。

### 4.3 tanh 函数

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

特点：

- 输出范围是 $(-1,1)$。
- 输出以 0 为中心，通常比 sigmoid 更适合作为隐藏层激活函数。

缺点：

- 输入绝对值较大时梯度会变小。
- 在较深网络中仍可能出现梯度消失。

### 4.4 ReLU 函数

$$
\text{ReLU}(x) = \max(0, x)
$$

特点：

- 输出范围是 $[0,+\infty)$。
- 计算简单，训练速度通常较快。
- 是隐藏层中非常常用的激活函数。

缺点：

- 当神经元长期输入小于等于 0 时，输出和梯度都为 0。
- 这种情况叫 Dead ReLU，神经元可能很难继续更新。

### 4.5 Leaky ReLU 函数

$$
\text{Leaky ReLU}(x) = \max(\alpha x, x)
$$

其中 $\alpha$ 是一个很小的正数，例如 $0.01$。

特点：

- 当 $x \le 0$ 时仍保留一个较小斜率。
- 可以缓解 Dead ReLU 问题。
- 计算仍然比较简单。

### 4.6 Softmax 函数

$$
\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{K} e^{x_j}}
$$

其中：

- $x_i$：第 $i$ 个类别的原始输出值
- $K$：类别总数
- $\text{softmax}(x_i)$：第 $i$ 个类别的预测概率

特点：

- 输出范围是 $(0,1)$。
- 所有类别概率之和为 1。
- 常用于多分类任务的输出层。

### 4.7 激活函数选择

| 位置 | 常见选择 | 说明 |
| --- | --- | --- |
| 隐藏层 | ReLU、Leaky ReLU、tanh | ReLU 最常用，训练速度通常较快 |
| 二分类输出层 | sigmoid | 输出一个正类概率 |
| 多分类输出层 | softmax | 输出各类别概率分布 |
| 回归输出层 | 线性输出 | 直接预测连续数值 |

简单来说：隐藏层主要考虑训练稳定性和表达能力，输出层主要根据任务类型选择。

## 5. 损失函数 Loss Function

### 5.1 定义

损失函数用于衡量模型预测值和真实值之间的差距。损失越小，说明模型当前预测越接近真实标签。

训练时，优化器会根据损失函数的梯度更新模型参数。

### 5.2 均方误差 MSE

均方误差（Mean Squared Error, MSE）常用于回归任务，计算预测值和真实值之间平方误差的平均值。

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

其中：

- $n$：样本数量
- $y_i$：第 $i$ 个样本的真实值
- $\hat{y}_i$：第 $i$ 个样本的预测值
- $(y_i - \hat{y}_i)^2$：单个样本的平方误差

有时也会写成 $\frac{1}{2n}$ 的形式，主要是为了求导时抵消平方项产生的系数 2。

### 5.3 二元交叉熵 Binary Cross-Entropy

二元交叉熵（Binary Cross-Entropy, BCE）常用于二分类任务，衡量预测概率和真实标签之间的差异。

$$
BCE = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]
$$

其中：

- $y_i \in \{0,1\}$：真实类别标签
- $\hat{y}_i$：模型预测为正类的概率
- $\log(\hat{y}_i)$：预测为正类时的对数概率
- $\log(1 - \hat{y}_i)$：预测为负类时的对数概率

直观理解：真实类别对应的预测概率越大，损失越小；真实类别对应的预测概率越小，损失越大。

### 5.4 多分类交叉熵 Cross-Entropy

多分类交叉熵常用于多分类任务，衡量预测类别分布和真实类别分布之间的差异。

$$
CE = -\frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{K} y_{ij} \log(\hat{y}_{ij})
$$

其中：

- $K$：类别数量
- $y_{ij}$：第 $i$ 个样本在第 $j$ 类上的真实标签，通常是 one-hot 编码
- $\hat{y}_{ij}$：模型预测第 $i$ 个样本属于第 $j$ 类的概率
- $CE$：多分类交叉熵损失

### 5.5 损失函数选择

| 任务 | 常用损失函数 | 输出层常见形式 |
| --- | --- | --- |
| 回归 Regression | MSE、MAE | 线性输出 |
| 二分类 Binary Classification | Binary Cross-Entropy | sigmoid |
| 多分类 Multi-Class Classification | Cross-Entropy | softmax |

选择损失函数时，要让输出层形式、标签格式和损失函数互相匹配。例如多分类任务通常使用 one-hot 标签或类别编号，并配合交叉熵损失。

## 6. 优化器 Optimizer

### 6.1 定义

优化器是用于更新神经网络参数的方法。它根据损失函数的梯度调整参数，使损失逐步降低。

基本更新形式为：

$$
\theta \leftarrow \theta - \eta \nabla_{\theta} L(\theta)
$$

其中：

- $\theta$：模型参数
- $L(\theta)$：损失函数
- $\eta$：学习率 Learning Rate
- $\nabla_{\theta} L(\theta)$：损失函数对参数的梯度

### 6.2 批量梯度下降 Batch Gradient Descent

批量梯度下降每次使用整个训练集计算梯度，梯度方向比较稳定，但数据量大时计算成本高、更新慢。

适合：数据量较小、希望梯度更稳定的场景。

### 6.3 随机梯度下降 Stochastic Gradient Descent

随机梯度下降（Stochastic Gradient Descent, SGD）每次使用一个样本计算梯度并更新参数。

$$
\theta \leftarrow \theta - \eta \nabla_{\theta} L(\theta; x_i, y_i)
$$

其中 $(x_i, y_i)$ 表示单个训练样本。

特点：

- 更新频繁，计算开销小。
- 梯度噪声较大，训练曲线可能不稳定。
- 有时噪声反而能帮助模型跳出较差的局部区域。

实际训练中更常用 Mini-batch Gradient Descent，即每次使用一小批样本计算梯度。

### 6.4 Momentum

Momentum 在更新参数时会累积过去的梯度方向，减少震荡，让参数更新更平滑。

直观理解：普通 SGD 只看当前梯度，Momentum 会参考之前的更新方向，类似给参数更新增加“惯性”。

### 6.5 Adam 优化器

Adam（Adaptive Moment Estimation）是一种自适应学习率优化算法，结合了 Momentum 和 RMSprop 的思想。

特点：

- 对不同参数自动调整学习率。
- 收敛速度通常较快。
- 是深度学习中非常常用的默认优化器。

### 6.6 优化器选择

| 优化器 | 特点 | 适合场景 |
| --- | --- | --- |
| Batch GD | 梯度稳定，但计算慢 | 小数据集、教学理解 |
| SGD | 简单，泛化能力常较好 | 需要细致调参的训练任务 |
| SGD + Momentum | 比普通 SGD 更稳定 | 图像分类等常见深度学习任务 |
| Adam | 自适应学习率，收敛快 | 初始实验、NLP、复杂模型 |

学习率通常比优化器本身更敏感：学习率过大可能震荡甚至发散，学习率过小会导致训练很慢。

## 7. 卷积神经网络 CNN

### 7.1 定义

卷积神经网络（Convolutional Neural Network, CNN）是一种擅长处理图像等网格结构数据的神经网络。它通过卷积层、激活函数、池化层和全连接层逐步提取特征，并完成分类、检测或回归任务。

CNN 的核心思想是：保留图像的空间结构，在局部区域中寻找有用模式，再把这些局部模式逐层组合成更高级的特征。

### 7.2 为什么不用普通全连接网络直接处理图像

图像具有明显的空间结构：相邻像素之间通常有关联，局部区域能组成边缘、角点、纹理、形状等特征。

如果直接使用全连接网络：

- 需要先把图像展平成一维向量，原来的上下左右关系会被削弱。
- 参数量会非常大，例如一张 $224 \times 224 \times 3$ 的图像就有 150528 个输入值。
- 模型不容易复用局部特征，也更容易过拟合。

CNN 通过局部连接和权重共享处理图像。局部连接让模型关注小区域，权重共享让同一个卷积核可以在整张图上寻找同一种模式。

### 7.3 卷积层：特征提取 Convolutional Layer

卷积层的主要作用是提取局部特征。可以把一个卷积核理解成一个“小型模式检测器”：

- 某个卷积核可能对竖直边缘响应强。
- 某个卷积核可能对水平边缘响应强。
- 某个卷积核可能对某种纹理或颜色变化响应强。

卷积核在图像上滑动时，每到一个位置，就计算这个局部区域和卷积核的匹配程度。匹配程度越高，输出值越大。所有位置的输出放在一起，就形成一张特征图（Feature Map）。

直观理解：

| 对象 | 可以理解成 |
| --- | --- |
| 卷积核 Kernel / Filter | 检测某种局部模式的工具 |
| 卷积运算 | 在每个位置检查这种模式是否出现 |
| 特征图 Feature Map | 记录这种模式在各个位置的响应强度 |
| 多个卷积核 | 同时检测多种不同模式 |

二维卷积可以简化写成：

$$
Y_{i,j} = \sum_{u=1}^{k_h} \sum_{v=1}^{k_w} W_{u,v} X_{i+u-1,j+v-1} + b
$$

其中：

- $X$：输入特征图
- $W$：卷积核权重
- $b$：偏置
- $k_h, k_w$：卷积核高度和宽度
- $Y_{i,j}$：输出特征图在位置 $(i,j)$ 的值

需要注意：卷积层不是简单把图像变小，而是在原图或上一层特征图上寻找有用模式。特征图的数值表示“某种特征在某个位置出现得明显不明显”。

### 7.4 特征如何逐层变抽象

CNN 的前几层通常学习简单特征，后面的层会把简单特征组合成更复杂的特征。

| 网络位置 | 常见特征 | 直观例子 |
| --- | --- | --- |
| 浅层卷积 | 边缘、颜色变化、简单纹理 | 横线、竖线、明暗变化 |
| 中间层卷积 | 局部形状、部件 | 眼睛、轮廓、角点 |
| 深层卷积 | 更抽象的语义特征 | 人脸、车辆、器官区域 |

所以“特征提取”不是一次完成的，而是逐层完成的：前面层提取基础模式，后面层把基础模式组合成更有意义的结构。

### 7.5 压缩：Stride、Pooling 和下采样

CNN 中的压缩主要是指降低特征图的空间尺寸，也叫下采样（Downsampling）。压缩的目的不是随便丢信息，而是在保留主要特征的同时减少计算量。

常见压缩方式：

| 方法 | 是否有可学习参数 | 作用 |
| --- | --- | --- |
| Stride > 1 的卷积 | 有 | 一边提取特征，一边降低尺寸 |
| Max Pooling | 无 | 保留局部区域中最强的特征响应 |
| Average Pooling | 无 | 保留局部区域的平均趋势 |

以 $2 \times 2$ 最大池化为例：

$$
\begin{bmatrix}
1 & 3 \\
2 & 5
\end{bmatrix}
\rightarrow 5
$$

它的含义是：这个局部区域里，最明显的特征响应是 5，所以用 5 代表这一小块区域。

池化层和卷积层的区别：

- 卷积层：通过可学习的卷积核提取新特征。
- 池化层：没有可学习参数，只对已有特征做压缩和筛选。
- Stride 卷积：既能学习特征，也能降低尺寸。

### 7.6 压缩为什么有用

压缩特征图有三个主要作用：

1. 减少计算量。特征图越小，后续卷积需要处理的位置越少。
2. 扩大感受野。后面的神经元能看到原图中更大的区域。
3. 提高一定的位置容忍度。特征稍微移动一点，池化后仍可能保留下来。

例如一条边缘在图像里稍微平移，如果每个像素位置都严格对应，模型会很敏感；池化后只保留局部区域的主要响应，可以让模型对小范围平移不那么敏感。

但压缩也有代价：

- 细节会减少。
- 位置信息会变粗。
- 过度池化可能让小目标或边界信息丢失。

所以 CNN 不是越压缩越好，而是在“保留有用特征”和“减少计算量”之间做平衡。

### 7.7 感受野 Receptive Field

感受野是指某一层特征图上的一个位置，能够“看到”原始输入图像中的多大区域。

直观理解：如果某个神经元最后判断“这里像猫耳朵”，它不是凭空判断的，而是根据原图中一块区域的信息判断的。这块区域就是它的感受野。

### 7.8 为什么感受野会变大

单个 $3 \times 3$ 卷积核一次只能看局部 $3 \times 3$ 区域，但多层卷积叠加后，后面的神经元能间接看到更大的原图区域。

例如连续使用两个 $3 \times 3$ 卷积：

1. 第一层的一个输出位置来自原图中的 $3 \times 3$ 区域。
2. 第二层的一个输出位置会看第一层中相邻的 $3 \times 3$ 个位置。
3. 第一层这些相邻位置各自又对应原图中的局部区域。
4. 合在一起，第二层的一个位置大约能看到原图中的 $5 \times 5$ 区域。

继续叠加卷积层，感受野会进一步变大：

| 层数 | 单个位置大致能看到的原图区域 |
| --- | --- |
| 1 个 $3 \times 3$ 卷积 | $3 \times 3$ |
| 2 个 $3 \times 3$ 卷积 | $5 \times 5$ |
| 3 个 $3 \times 3$ 卷积 | $7 \times 7$ |

这里假设 stride 为 1，padding 不改变空间对齐。实际网络里如果加入 stride、pooling 或 dilation，感受野增长会更快。

### 7.9 感受野和特征提取的关系

感受野小的时候，模型更关注局部细节；感受野大的时候，模型能结合更大范围的上下文。

| 感受野大小 | 更容易提取的特征 |
| --- | --- |
| 小感受野 | 边缘、角点、简单纹理 |
| 中等感受野 | 局部形状、物体部件 |
| 大感受野 | 整体结构、语义区域 |

所以 CNN 通常前面层关注局部细节，后面层关注更大范围的结构。不是因为后面层“天然更聪明”，而是因为后面层的感受野更大，并且输入已经是前面层提取过的特征。

### 7.10 感受野和压缩的关系

Stride 和 Pooling 会降低特征图尺寸，同时让后面层的一个位置对应原图中更大的区域。

可以这样理解：

- stride 为 1：卷积核一步一步滑动，空间信息保留更细。
- stride 大于 1：卷积核跳着滑动，输出更小，后面每个位置覆盖的原图范围更大。
- pooling：把局部区域压缩成一个值，减少位置数量，让后面层更快获得较大的观察范围。

压缩有助于扩大感受野，但也会让位置信息变粗。因此图像分类可以较多使用下采样，分割、检测这类需要精确位置的任务则要更谨慎地压缩。

### 7.11 输出尺寸和参数量

卷积层输出尺寸常用公式为：

$$
H_{out} = \left\lfloor \frac{H_{in} + 2P - K}{S} \right\rfloor + 1
$$

$$
W_{out} = \left\lfloor \frac{W_{in} + 2P - K}{S} \right\rfloor + 1
$$

其中：

- $H_{in}, W_{in}$：输入特征图的高和宽
- $K$：卷积核大小，若高宽不同可分别写成 $K_h, K_w$
- $P$：Padding 大小
- $S$：Stride 步长
- $H_{out}, W_{out}$：输出特征图的高和宽

卷积层参数量为：

$$
K_h \times K_w \times C_{in} \times C_{out} + C_{out}
$$

其中：

- $C_{in}$：输入通道数
- $C_{out}$：输出通道数，也就是卷积核个数
- 最后的 $C_{out}$：每个输出通道对应一个偏置

这个公式能看出：卷积层参数量和图像高宽无直接关系，而主要和卷积核大小、输入通道数、输出通道数有关。这就是 CNN 比直接使用全连接层更适合处理图像的重要原因。

### 7.12 全连接层 Fully Connected Layer

CNN 前面的卷积层和池化层负责提取、筛选和压缩特征，后面的全连接层通常负责把提取到的特征映射到最终输出空间。

常见流程为：

$$
\text{Image} \rightarrow \text{Conv} \rightarrow \text{ReLU} \rightarrow \text{Pooling} \rightarrow \text{Flatten} \rightarrow \text{FC} \rightarrow \text{Output}
$$

其中：

- Conv：提取局部特征
- ReLU：引入非线性，保留有效响应
- Pooling：降低空间尺寸，筛选主要响应
- Flatten：把特征图展开成向量
- FC：完成分类或回归输出

### 7.13 一个完整直观例子

假设输入是一张猫的图片：

1. 第一层卷积可能检测边缘、颜色变化和简单纹理。
2. 池化层把局部最明显的响应保留下来，压缩特征图尺寸。
3. 更深的卷积层把边缘和纹理组合成眼睛、耳朵、胡须等局部结构。
4. 最后几层把这些局部结构组合成“猫”这个整体判断。

所以可以把 CNN 理解成：

$$
\text{像素} \rightarrow \text{边缘/纹理} \rightarrow \text{局部部件} \rightarrow \text{整体类别}
$$

### 7.14 注意点

- 卷积核大小影响感受野，常见大小有 $3 \times 3$、$5 \times 5$。
- Padding 会影响输出特征图尺寸，常用于保留边缘信息。
- 池化能减少计算量，但过度池化可能损失细节。
- Stride 变大也能压缩尺寸，但可能跳过一些细节。
- 图像任务中常配合数据增强，例如随机裁剪、翻转、颜色扰动等。
- CNN 适合图像任务，但训练时仍要关注过拟合、学习率和数据分布差异。

## 8. 训练流程和常见问题 Training Workflow

### 8.1 数据划分

深度学习训练通常会把数据分成训练集、验证集和测试集。

| 数据集 | 作用 |
| --- | --- |
| 训练集 Training Set | 用来更新模型参数 |
| 验证集 Validation Set | 用来调参、选择模型和观察过拟合 |
| 测试集 Test Set | 用来评估最终模型效果 |

不能用测试集反复调参，否则测试结果会变得不客观。

### 8.2 Epoch、Batch 和 Iteration

| 概念 | 含义 |
| --- | --- |
| Batch | 一次送入模型训练的一小批样本 |
| Iteration | 使用一个 batch 完成一次参数更新 |
| Epoch | 模型完整看完一遍训练集 |

例如训练集有 10000 张图片，batch size 为 100，则 1 个 epoch 包含 100 次 iteration。

### 8.3 过拟合和欠拟合

过拟合是指模型在训练集上表现很好，但在验证集或测试集上表现较差。欠拟合是指模型连训练集都学不好。

| 问题 | 表现 | 常见处理 |
| --- | --- | --- |
| 过拟合 Overfitting | 训练损失低，验证损失高 | 增加数据、数据增强、正则化、Dropout、早停 |
| 欠拟合 Underfitting | 训练损失和验证损失都高 | 增大模型、训练更久、调学习率、改进特征 |

判断模型问题时不要只看训练集损失，要同时看验证集曲线。

### 8.4 Dropout

Dropout 是一种常见正则化方法。训练时随机让一部分神经元暂时失效，减少神经元之间过强的依赖关系。

直观理解：模型不能只依赖少数几个神经元，必须学到更稳健的特征。

注意点：

- Dropout 只在训练时启用，推理时关闭。
- Dropout 比例过大可能导致模型学不充分。
- CNN 中不一定每一层都需要 Dropout，常见于全连接层附近。

### 8.5 Batch Normalization

Batch Normalization（BN）会对一层的中间输出做标准化，使训练更稳定。

常见作用：

- 缓解训练过程中分布变化带来的不稳定。
- 允许使用稍大的学习率。
- 在一定程度上有正则化效果。

常见位置是卷积层或全连接层之后、激活函数之前或之后，具体写法会随框架和模型结构变化。

### 8.6 梯度消失和梯度爆炸

梯度消失是指反向传播时梯度越来越小，前面层参数很难更新。梯度爆炸是指梯度过大，导致训练不稳定甚至损失变成异常值。

常见处理方法：

- 使用 ReLU、Leaky ReLU 等更合适的激活函数。
- 合理初始化参数。
- 使用 Batch Normalization。
- 调小学习率。
- 对梯度爆炸可以使用梯度裁剪 Gradient Clipping。

### 8.7 学习率调整

学习率决定每次参数更新的步子大小。

| 情况 | 可能原因 | 调整方向 |
| --- | --- | --- |
| 损失震荡或发散 | 学习率过大 | 降低学习率 |
| 损失下降很慢 | 学习率过小 | 适当提高学习率 |
| 前期下降后期停滞 | 固定学习率不合适 | 使用学习率衰减 |

常见策略包括 Step Decay、Cosine Annealing、Warmup 等。复习时先理解核心思想：训练初期可以走得快一些，后期需要更小步地靠近较优解。

### 8.8 一个基本训练流程

1. 准备数据，并划分训练集、验证集和测试集。
2. 定义模型结构，例如 MLP 或 CNN。
3. 选择损失函数和优化器。
4. 设置学习率、batch size、epoch 等超参数。
5. 在训练集上训练，在验证集上观察效果。
6. 根据验证结果调整模型和超参数。
7. 最后只在测试集上做一次相对客观的评估。
